# **학교 시리즈 ( 2013~2014)**

## 2 COOL 4 SKOOL
* 7곡

```python
url = 'https://www.melon.com/album/detail.htm?albumId=2188611'
driver.get(url)

# 사이트 부담을 줄이기 위해 똑같은 데이터가 수집되는 컬럼은 수집 대신 
release_date = '2013.06.12'
artist = '방탄소년단'
album_name = "2 COOL 4 SKOOL"
genre = '랩/힙합, 발라드'

song_name = []
like_count = []
lyrics = []
for i in range(1,8):

    # 곡 정보 클릭
    driver.find_element_by_xpath(f'//*[@id="frm"]/div/table/tbody/tr[{i}]/td[3]/div/a').click()
    time.sleep(1)
    
    # 1. 곡 제목 수집
    song_name.append(driver.find_element_by_xpath('//*[@id="downloadfrm"]/div/div/div[2]/div[1]/div[1]').text)
    
    # 2. 곡 좋아요 수
    like_count.append(driver.find_element_by_xpath('//*[@id="d_like_count"]').text)
   
    # 3. 가사 수집
   
    lyrics.append(driver.find_element_by_xpath('/html/body/div[1]/div[3]/div/div/div/div[2]/div[2]/div').text)

    # 뒤로가기
    driver.back()
```

```python
df1 = pd.DataFrame({'release_date' : release_date,
                     'artist' : artist,
                    'series' :'school',
                     'album_name' : album_name,
                     'song_name' : song_name,
                     'genre' : genre,
                     'like_count' : like_count,
                     'lyrics' : lyrics})
df1.head()
```

```python
# 리스트를 데이터프레임으로
df12 = pd.DataFrame({'release_date' : release_date,
                     'artist' : artist,
                     'series' :'love_yourself',
                     'album_name' : album_name,
                     'song_name' : song_name,
                     'genre' : genre,
                     'like_count' : like_count,
                     'lyrics' : lyrics})
df12
```
